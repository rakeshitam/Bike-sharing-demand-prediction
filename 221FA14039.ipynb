```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd # import pandas

# Sample data (replace with your actual data)
data = {'Rented Bike Count': [500, 1000, 1500, 2000, 2500, 3000, 3500],
        'Temperature(°C)': [-20, -10, 0, 10, 20, 30, 40]}

df = pd.DataFrame(data) # create a dataframe

# Now you can use df
X = df['Rented Bike Count'].values.reshape(-1, 1)  # Reshape for sklearn
y = df['Temperature(°C)'].values
```

```python
# Create a linear regression model
model = LinearRegression()
model.fit(X,y)
```

```python
# Predict y values based on the model
y_pred = model.predict(X)
```

```python
# Plot the data and the regression line
plt.scatter(X, y, label='Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel('Rented Bike Count') # changed label to match dataframe
plt.ylabel('Temperature(°C)') # changed label to match dataframe
plt.title('Linear Regression: Rented Bike Count vs Temperature(°C)') # changed title to match dataframe
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd

# Sample data (replace with your actual data)
data = {'Rented Bike Count': [500, 1000, 1500, 2000, 2500, 3000, 3500],
        'Temperature(°C)': [-20, -10, 0, 10, 20, 30, 40]}

df = pd.DataFrame(data)

# Now you can use df
X = df['Rented Bike Count'].values.reshape(-1, 1)
y = df['Temperature(°C)'].values

# Create a linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict y values based on the model
y_pred = model.predict(X)

# Calculate R-squared
from sklearn.metrics import r2_score
r_squared = r2_score(y, y_pred)

print("R-squared:", r_squared)
```
